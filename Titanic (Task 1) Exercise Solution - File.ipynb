> ## a-) Perform the following column subsetting manipulaitons on the dataset

### 1-) Only return *alone, alive and sex* column, with all rows 
Since here we are given the *names* of the columns , it is clear that we have to use the **loc** indexer instead of the **iloc** indexer. pay close attention to the syntax of the indexer.

```python
titanic.loc[:,['alone','alive','sex']]
```

### 2-) Only return columns from start to *class* , with all rows
Once again since the name of the column **class** is given we are bound to use the loc indexer. However we are performing the slicing operation instead of simple column subsetting. Dont worry! loc indexers can handle the slicing operations well.

```python
titanic.loc[:,:'class']
```

### 3-)  Only return top 10 rows and 2nd, 5th and 6th column
In the part , for both rows and columns only position of the columns and rows are give. Hence we will utlize the **iloc** indexer. Notice that this operation wants you to slice on rows and sub-set on the columns.

```python
titanic.iloc[:10,[2,5,6]]
```

> ## b-) Perform the following row subsetting manipulaitons on the dataset
In this section , you will combine boolean indexing operations along with the loc indexer. Please note that iloc idexer fall short 

### 1-) Only Return records where the passenger was alone , all columns 

```python
titanic.loc[titanic['alone']==True, :]
```

### 2-) Only return records where the passenger was alone and did not survive the crash, all columns
Multiple condition check . Notice rthe square brackets and & operator. 

```python
titanic.loc[(titanic['alone']==True) & (titanic['alive']=='no') ,:]
```

### 3-) Only return the column 'fare' for female passengers who were in 1st class of the ship 
Notice multiple conditions and retrieval of only the 'fare' column

```python
titanic.loc[(titanic['sex']=='female') & (titanic['class']=='First') , 'fare']
```
